```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import  date
```

```python
df = pd.read_csv("../data/data.csv")
df.columns = [col.replace(" ","_").strip().lower() for col in df.columns]
df.head()
```

# Categorical Features

```python
cat_col = df.select_dtypes("object")
cat_col_name = cat_col.columns.to_list()
cat_col.head()
```

```python
for col in cat_col.columns:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df[cat_col_name]
```

make,

```python
top_9_make = df["make"].value_counts().nlargest(9).index
df["make"] = df["make"].apply(lambda x: x if x in top_9_make else "other") 
top_9_model = df["model"].value_counts().nlargest(9)
df["model"] = df["model"].apply(lambda x: x if x in top_9_model else "other") 

catagery = list(set(",".join(df["market_category"]. unique().tolist()).split(",")))
for cat in catagery:
    df[cat] = df["market_category"].apply(lambda x:  1 if cat in x else 0)
df.columns  =  [col.replace(" ","_").strip().lower() for col in df.columns]
df.drop(columns="market_category",inplace=True)

top_9_style = df["vehicle_style"].value_counts().nlargest(9).index
df["vehicle_style"] =  df["vehicle_style"].apply(lambda x: x if x in top_9_style else "other")
```

```python
df.select_dtypes("object")
```

# numerical columns

```python
num_col = df.select_dtypes(exclude="object").columns.to_list()
 
```

```python


for col in num_col:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])
```
