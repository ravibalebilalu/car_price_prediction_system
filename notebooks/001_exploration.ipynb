```python
cat_col = df.select_dtypes("object")
cat_col.head()
```

## missing values columns

```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        print(col)
```

## categorical columns with more than 10 unique value counts

```python
for col in cat_col.columns:
    if df[col].nunique() > 10:
        print(col)
```
